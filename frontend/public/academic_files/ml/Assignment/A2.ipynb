#Training a baseline model
There are many approaches for training a sequence classification model for text data. In this lab, we're giving you code that mirrors what you find if you look up how to [train a text classifier](https://scikit-learn.org/stable/auto_examples/text/index.html), where we'll train an SVM on tf-idf features (numeric representations of each text field based on word occurrences).

```python
# Final Question: What is the dif between technovation and vectorization?
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
```

```python
# Original Way
classifier1 = Pipeline([
    ('vect', TfidfVectorizer()),
    ('clf', LinearSVC())
])

classifier = classifier1.fit(train['review'], train['label'])
```

```python
# If i don't want to use Pipeline
# Equivalent to the above block

vectorizer = TfidfVectorizer()
train_vectors = vectorizer.fit_transform(train['review'])
test_vectors = vectorizer.transform(test['review'])

classe = LinearSVC()
classe.fit(train_vectors, train['label'])
```

# Evaluating model accuracy


```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
def evaluate (model):
  predicted = model.predict(test['review'])
  print(metrics.classification_report(test['label'], predicted))
  print("Accuracy:", accuracy_score(test['label'], predicted))
```

```python
evaluate(classifier)
# How many was correct: precision
# How many is predicted on the whole thing: recall
# How many instance where good and bad: support
# The overall accuracy is: Accuracy
```
