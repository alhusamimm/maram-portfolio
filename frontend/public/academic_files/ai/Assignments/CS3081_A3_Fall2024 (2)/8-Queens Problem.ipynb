```python
import random

# Fitness Function
def fitness(board):
    conflicts = 0
    for i in range(len(board)):
        for j in range(i+1, len(board)):
            if board[i] == board[j] or abs(board[i] - board[j]) == j - i:
                conflicts += 1
    return 28 - conflicts  # Max conflicts is 28, so higher fitness is better

# Selection: Fitness-based selection using roulette wheel
def select(population, fitness_values):
    total_fitness = sum(fitness_values)
    selection_probs = [f / total_fitness for f in fitness_values]
    selected_index = random.choices(range(len(population)), selection_probs)[0]
    return population[selected_index]

# Crossover: Single-point crossover
def crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    child = parent1[:point] + parent2[point:]
    return child

# Mutation: Random mutation of a single queen's position
def mutate(board):
    index = random.randint(0, len(board) - 1)
    new_position = random.randint(0, len(board) - 1)
    board[index] = new_position
    return board

# Main Genetic Algorithm
def genetic_algorithm(population_size, generations, mutation_rate):
    population = [random.sample(range(8), 8) for _ in range(population_size)]
    generation = 0
    
    while generation < generations:
        fitness_values = [fitness(individual) for individual in population]
        
        # Check for solution
        if max(fitness_values) == 28:  # A solution is found (zero conflicts)
            solution_index = fitness_values.index(max(fitness_values))
            solution = population[solution_index]
            return solution
        
        # Select parents
        new_population = []
        for _ in range(population_size):
            parent1 = select(population, fitness_values)
            parent2 = select(population, fitness_values)
            
            # Crossover
            child = crossover(parent1, parent2)
            
            # Mutation
            if random.random() < mutation_rate:
                child = mutate(child)
            
            new_population.append(child)
        
        population = new_population
        generation += 1
        
    # If no solution found within max generations, return the best solution
    best_solution = population[fitness_values.index(max(fitness_values))]
    return best_solution

# Display the board configuration
def display_board(board):
    for row in range(8):
        line = ""
        for col in range(8):
            if board[col] == row:
                line += "Q "
            else:
                line += ". "
        print(line)

# Run the Genetic Algorithm
population_size = 100
generations = 1000
mutation_rate = 0.05

solution = genetic_algorithm(population_size, generations, mutation_rate)
print("Solution found:")
display_board(solution)
```
